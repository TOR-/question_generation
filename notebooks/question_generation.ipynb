## Single task QA

```
%cd question_generation
```

```
from pipelines import pipeline
```

```
nlp_qg_base = pipeline("question-generation", model="valhalla/t5-base-qg-hl")
nlp_qg_small = pipeline("question-generation", model="valhalla/t5-small-qg-hl")
```

```
qqg = [(nlp_qg_base(t), nlp_qg_small(t)) for t in texts]
```

```
for i, qq in enumerate(qqg):
  print(f"{'='*25}\n= TEXT   : {texts[i][:12]} =")
  for (bq,sq) in itertools.zip_longest(*qq, fillvalue=None):
    print(f"/BASE  Qn: {bq['question']}\n\\========: {bq['answer']}")
    print(f"/SMALL Qn: {sq['question']}\n\\========: {sq['answer']}")
```
